```python
import numpy as np

def hitung_probeam_specs(energi_mev):
    # Konstanta Hasil Fitting dari Jupyter Anda
    a0 = 2.60137049
    a1 = 1467.402324
    b0 = 1.13883335
    massa_proton = 938.272 # MeV/c^2
    
    # Perhitungan Relativistik
    p = np.sqrt(energi_mev**2 + 2 * energi_mev * massa_proton)
    beta = p / (energi_mev + massa_proton)
    
    # 1. Beam Position Spread (mm)
    sigma_0 = np.sqrt(a0**2 + (a1 / p)**2)
    
    # 2. Beam Angular Spread (mrad)
    sigma_p = (b0 / (beta * p)) * 1000 # dikali 1000 untuk konversi rad ke mrad
    
    return sigma_0, sigma_p

# Contoh penggunaan untuk daftar energi simulasi Anda
daftar_energi = [150.000000,
149.044962,
148.396135,
147.745115,
147.091879,
146.436402,
145.778657,
145.118620,
144.456263,
143.791559,
143.124480,
142.454999,
141.783086,
141.108713,
140.431848,
139.752463,
139.070524,
138.386001,
137.698861,
137.009070,
136.316595,
135.621400,
134.923451,
134.222710,
133.519141,
132.812706,
132.103365,
131.391080,
130.675808,
129.957510,
129.236141,
128.511658,
127.784016,
127.053170,
126.319073] 

print("Spesifikasi Varian ProBeam untuk File TOPAS Anda:")
print("-" * 55)
for T in daftar_energi:
    pos_spread, ang_spread = hitung_probeam_specs(T)
    print(f"Energi: {T:5.1f} MeV | Pos Spread: {pos_spread:5.3f} mm | Ang Spread: {ang_spread:5.2f} mrad")
```
